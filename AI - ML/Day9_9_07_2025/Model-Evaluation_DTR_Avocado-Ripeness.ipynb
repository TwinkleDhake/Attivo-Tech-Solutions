```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:\Attivo Trainning\Technical Videos\AI_ML\Jupyter Notebook Daily Practice\Day9_9_07_2025\avocado_ripeness_dataset.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
col_to_encode = ["color_category", "ripeness"]
```

```python
for columns in col_to_encode:
    df[columns] = le.fit_transform(df[columns])
    df.head()
```

```python
X = df.drop(['sound_db', 'size_cm3', 'ripeness'], axis = 'columns')
```

```python
#X = df.name
```

```python
X.head()
```

```python
y = df['ripeness']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size = 0.3, random_state = 3)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
reg = DecisionTreeRegressor(random_state = 2)
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(100,100))
plot_tree(reg,feature_names = X.columns, filled = True)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
print (mae)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
