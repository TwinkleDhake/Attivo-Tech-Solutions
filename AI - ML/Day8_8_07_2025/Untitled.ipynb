```python
Dt = df['Date']
Gr_Inc = df['gross income']
```

```python
plt.figure(figsize=(30,10))
plt.scatter(Dt, Gr_Inc, color = 'green', marker = 'o')
plt.title("Super Market Sales - Gross Income per Unit Price")
plt.xlabel("Date - Independent variable")
plt.ylabel("Gross Income - dependent variable")
plt.xticks(rotation=45, ha='right')
plt.grid(axis = 'y')
plt.tight_layout()
plt.show()
```
