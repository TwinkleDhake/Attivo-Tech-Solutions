```python
x = [10, 20, 30, 40]
y = [2, 4, 8, 16]
```

```python
plt.scatter(x, y, color = 'green', marker = 'o')
plt.title("Scatter plot Visualization")
plt.xlabel("x - Independent variable")
plt.ylabel("y - dependent variable")
plt.show
```
