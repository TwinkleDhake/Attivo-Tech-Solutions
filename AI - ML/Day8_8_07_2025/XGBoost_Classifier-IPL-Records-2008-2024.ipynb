```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
col_to_encode = ["Teams", "Venue", "Toss_Winner", "Toss_Decision", "Match_Winner", "Win_Type", "Player_of_Match", "Umpire", "Umpire1", "Umpire2"]
```

```python
for columns in col_to_encode:
    df[columns] = le.fit_transform(df[columns])
    df.head()
```

```python
X = df.drop('Toss_Decision', axis = 'columns')
```

```python
X.head()
```

```python
y = df['Toss_Decision']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)
```

```python
for df_ in [X_train, X_test]:
    df_['Date'] = pd.to_datetime(df_['Date'])
    df_['year'] = df_['Date'].dt.year
    df_['month'] = df_['Date'].dt.month
    df_['day'] = df_['Date'].dt.day
    df_.drop('Date', axis=1, inplace=True)
```

```python
import xgboost as xgb
```

```python
clf = xgb.XGBClassifier(eval_metric = 'logloss')
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```
