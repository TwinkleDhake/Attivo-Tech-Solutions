```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
y = [120, 130, 90, 20, 200]
```

```python
plt.plot(x, y, linestyle = '-', marker = '*', color = 'green')
plt.title("Sales of a product")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
