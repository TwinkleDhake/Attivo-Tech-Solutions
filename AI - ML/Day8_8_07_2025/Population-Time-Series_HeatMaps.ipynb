```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
```

```python
df.drop(['realtime_start', 'date', 'realtime_end'], axis = 'columns', inplace = True)
```

```python
df.head()
```

```python
pivot_df = df.pivot_table(values = 'value', index = 'year', aggfunc = 'mean')
plt.figure(figsize = (30,10))
sns.heatmap(pivot_df, annot = True, cmap = 'YlGnBu', fmt = '.0f') #, char = True)
plt.title("Population Heatmap")
plt.xlabel("Population")
plt.tight_layout()
plt.show()
```
