```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\yasha\Desktop\AI_ML\AI_ML_DailyPractice\Day6_4_07_2025\Car Dekho\cardekho_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
col_to_encode = ["Fuel_Type", "Seller_Type", "Transmission"]
```

```python
for columns in col_to_encode:
    df[columns] = le.fit_transform(df[columns])
    df.head()
```

```python
X = df.drop(['Car_Name', 'Selling_Price'], axis = 'columns')
```

```python
X.head()
```

```python
y = df['Selling_Price']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size = 0.3, random_state = 4)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
reg = DecisionTreeRegressor(random_state = 2)
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(100,100))
plot_tree(reg,feature_names = X.columns, filled = True)
plt.show()
```
