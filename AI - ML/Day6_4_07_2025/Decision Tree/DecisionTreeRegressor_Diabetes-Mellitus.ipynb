```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
col_to_encode = ["Gender", "Polyuria", "Polydipsia", "sudden weight loss", "weakness", "Polyphagia", "Genital thrush", "visual blurring", "Itching", "Irritability", "delayed healing",	"partial paresis",	"muscle stiffness",	"Alopecia",	"Obesity",	"class"]
```

```python
for columns in col_to_encode:
    df[columns] = le.fit_transform(df[columns])
    df.head()
```

```python
X = df.drop(['Genital thrush', 'visual blurring', 'Itching', 'Irritability'], axis = 'columns')
```

```python
#X = df.name
```

```python
X.head()
```

```python
y = df['Gender']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test, = train_test_split(X, y, test_size = 0.3, random_state = 3)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
reg = DecisionTreeRegressor(random_state = 2)
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```
