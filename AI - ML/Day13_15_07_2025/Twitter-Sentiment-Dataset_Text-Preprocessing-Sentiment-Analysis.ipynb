```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
```

```python
# Removal of @mentions
no_mentions = lwr_cs.apply(lambda x: re.sub(r'@\w+','', str(x)))
no_mentions
```

```python
# Removal of URLs
no_url = no_mentions.apply(lambda x: re.sub(r'https?://\S+|www\.\S+','',x))
no_url
```

```python
# Removal of Hashtags
no_hashtags = no_url.apply(lambda x: re.sub(r'#\S+','',x))
no_hashtags
```

```python
# Removal of Numbers
no_numbers = no_hashtags.apply(lambda x: re.sub(r'\d+','',x))
no_numbers
```

```python
','.join(stopwords.words('English'))
```

```python
tokenized = no_numbers.apply(lambda x: word_tokenize(x, preserve_line=True))
print(tokenized)
```

```python
# stop_word Removal
stop_words = set(stopwords.words('english'))
filtered_tokens = tokenized.apply(lambda tokens:[token for token in tokens if token not in stop_words and len(token)>1])    
print(filtered_tokens)
```

```python
# Stemming
stemmer = PorterStemmer()
stemmed_tokens = filtered_tokens.apply(lambda tokens: [stemmer.stem(token) for token in tokens])
stemmed_tokens
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = stemmed_tokens.apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
lemmatized_tokens
```

```python
final_cleaned_text = lemmatized_tokens.apply(lambda tokens: ' '.join(tokens))
```

```python
source_df['clean_text'] = final_cleaned_text
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from textblob import TextBlob
```

```python
#nltk.download('vader_lexicon')
```

```python
#VADER Sentiment Analysis
sia = SentimentIntensityAnalyzer()
```

```python
def get_vader_sentiment(text):
    score = sia.polarity_scores(str(text))
    if score['compound'] >= 0.05:
       return 'positive'
    elif score['compound'] <= -0.05:
       return 'negative'
    else:
       return 'neutral'
```

```python
source_df['Sentiment'] = source_df['clean_text'].apply(get_vader_sentiment)
```

```python
source_df.head()
```

```python
source_df.to_csv('VADER_sentiment_analysis', index = False)
```

```python
# TextBlob Sentiment Analysis
def get_textblob_sentiment(text):
    analysis = TextBlob(str(text))
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python
source_df['TB_Sentiment'] = source_df['clean_text'].apply(get_textblob_sentiment)
source_df.head()
```

```python
# Sentiment WordCloud
sentiments = ['positive', 'neutral', 'negative']
plt.figure(figsize = (25,10))
for i, sentiment in enumerate(sentiments,1):
    text = ' '.join(source_df[source_df['Sentiment'] == sentiment]['clean_text'].dropna())
    wordcloud = WordCloud(background_color = 'white', 
                          colormap = 'viridis' if sentiment == 'positive'
                                               else 'Blues' if sentiment == 'neutral'
                                               else 'Reds').generate(text)
    plt.subplot(1,3,i)
    plt.imshow(wordcloud, interpolation = 'bilinear')
    plt.title('Sentiment Analysis WordCloud', fontsize=20)
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.subplot(1,2,1)
source_df['TB_Sentiment'].value_counts().plot(kind = 'bar', color =['green', 'blue', 'red'])
plt.title('TextBlob Sentiment Analysis Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
