```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
col_to_encode = ["CarName",	"fueltype",	"aspiration", "doornumber", "carbody", "drivewheel", "enginelocation", "cylindernumber", "fuelsystem", "enginetype"]
car_prices_encoded = pd.get_dummies(car_prices, columns =col_to_encode, drop_first=True)
for columns in col_to_encode:
    car_prices[columns] = le.fit_transform(car_prices[columns])
    car_prices.head()
```

```python
X = car_prices_encoded.drop("price", axis=1)
```

```python
X.head()
```

```python
y = car_prices_encoded["price"]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 5)
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```
