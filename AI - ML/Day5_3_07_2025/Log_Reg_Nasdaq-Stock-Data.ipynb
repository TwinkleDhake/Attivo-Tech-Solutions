```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
threshold = 0.3
```

```python
#nsd.dropna(inplace=True)
```

```python
nsd = nsd.loc[:, nsd.isnull().mean() < threshold]
```

```python
num_cols = nsd.select_dtypes(include=['float64', 'int64']).columns
```

```python
nsd[num_cols] = nsd[num_cols].fillna(nsd[num_cols].median())
```

```python
cat_cols = nsd.select_dtypes(include=['object']).columns
```

```python
nsd[cat_cols] = nsd[cat_cols].fillna('Unknown')
```

```python
dummy_columns = ["symbol", "security_name", "zip", "listing_exchange", "etf", "financial_status", "zip", "sector", "city", "state", "revenueQuarterlyGrowth", "floatShares", "enterpriseValue", "country", "industry"]
```

```python
dummy_cols = [col for col in dummy_columns if col in nsd.columns]
```

```python
nsd = pd.get_dummies(nsd, columns=dummy_cols, drop_first=True)
```

```python
def categorize_dividend(value):
    if value == 0:
        return "No"
    elif value <= 0.5:
        return "Low"
    elif value <= 2:
        return "Medium"
    else:
        return "High"
nsd['dividend_category'] = nsd['latest_dividend'].apply(categorize_dividend)
```

```python
X = nsd.drop(['latest_dividend', 'dividend_category'], axis=1)
```

```python
X.head()
```

```python
y = nsd['dividend_category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```
